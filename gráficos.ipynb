```python
import pandas
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
# Gráfico sobre quantidade de pessoas por curso
mulheres = [1011,350,431,1983,1334,207,530]
homens = [7397,730,3022,9919,8302,2273,2673]

# Definindo a largura das barras
barWidth = 0.25

# Aumentando o gráfico
plt.figure(figsize=(12,3))

# Definindo a posição das barras
r1 = np.arange(len(homens))
r2 = [x + barWidth for x in r1]

 
# Criando as barras
plt.bar(r1, mulheres, color='lightskyblue', width=barWidth, label='Feminino')
plt.bar(r2, homens, color='lightcoral', width=barWidth, label='Masculino')

 
# Adiciando legendas as barras
plt.xlabel('Cursos')
plt.xticks([r + barWidth for r in range(len(homens))], ['CompBacharel', 'CompLicenciatura', 'SistInformação', 'Eng.Computação', 'Tads',
'Tec. Redes', 'GestãoTI'])
plt.ylabel('Quantidade')
plt.title('Quantidade de pessoas por curso')
 
# Criando a legenda e exibindo o gráfico
plt.legend()
plt.show()
```
